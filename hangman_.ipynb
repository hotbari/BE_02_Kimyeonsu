```python
import random

words = ['melon','banana','bluberry','tengerin','peach']
###행맨 클래스 만들자!
class HangmanGame :
    ##필요한 요소:정답 단어, 추측 단어, 시도 횟수 (기본적인 게임 세팅)
    def __init__(self):
        #단어 중 랜덤 선택
        self.word = random.choice(words)
        self.guesses = set()
        self.attempts = 6

    def display(self) : ##단어 현상태 표시
        result = "" #빈 문자열 준비

        for char in self.word : #문자 입력 받았을 때 정답에 있는지
            if char in self.guesses :
                result += char #있으면 그 문자를 출력
            else :
                result += '_' #없으면 _ 출력

        return result #만약 banana인데 a를 받았으면 _a_a_a 로 나옴


    ##맞추는 과정...
    def play(self):

        while self.attempts > 0 : #시도횟수가 남았으면 반복 실행
            print(hangman_pics[6 - self.attempts]) #인덱싱 활용
            print(self.display()) #단어 현상태 표시(method)

            guess = input('궁금한 알파벳을 입력하세요: ').lower() #추측 알파벳 입력 받음


            if guess in self.guesses : #중복이면
                print('중복입니다.')

            elif guess in self.word : #문자안에 있으면
                self.guesses.add(guess)
                print('단어를 맞추셨군요. 더 힘내봐요!')

                if set(self.word).issubset(self.guesses): #다 맞추면 - 추측한 글자, set 집합에 추가 겹치면 안되니까 set
                    print(f'정답입니다! {self.word}')
                    break



            else: #중복은 아닌데 문자 안에 없으면
                self.attempts -= 1
                print(f'틀렸습니다. 다시 시도해주세요. (남은 횟수: {self.attempts})')

        if self.attempts == 0 : #시도할 수 있는 횟수가 끝났으면
            print(hangman_pics[-1])
            print(f'게임 종료. 정답은: {self.word}')



#직접 실행된 모듈이라 __main__ 의 이름을 갖고
#파일에서 실행된거는 파일명임

# if __name__ == "__main__":
    # game = HangmanGame()
    # game.play()

game = HangmanGame()
game.play()
```
